```python
import pandas as pd
import yfinance as yf
```

```python
# Função para calcular o market cap
def calcular_market_cap(symbol):
    print(symbol)
    try:
        stock = yf.Ticker(f'{symbol}.SA')
        market_cap = stock.info.get("marketCap")
        return market_cap
    except:
        return None
```

```python
# https://www.dadosdemercado.com.br/bolsa/acoes
df = pd.read_csv('acoes-listadas-b3.csv')
```

```python
df['marketCap'] = df['Ticker'].apply(calcular_market_cap)
```

```python
df['marketCap_date'] = '17/02/2024'
df = df.sort_values('marketCap', ascending=False)
```

```python
df.drop_duplicates(subset='Nome').head(20).reset_index(drop=True)
```
